```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import pandas as pd
import numpy as np

pd.options.display.max_columns=8
```

## Reading the Images

```python
image_names = ['baboon.png', 'flowers.png', 'graffiti.jpg', 'lena.png', 'umbrella.jpg']
```

```python
images = []
for image_name in image_names:
    images.append(mpimg.imread(image_name))
```

## Visualizing the Images

```python
def print_image_dimensions(image):
    for index, dimension in enumerate(np.rollaxis(image, 2)):
        print('Dimension {}:\n'.format(index+1), dimension, '\n')
```

```python
fig = plt.figure(figsize=(15, 15))
columns = 2
rows = 3
for index, image in enumerate(images):
    fig.add_subplot(rows, columns, index+1)
    plt.imshow(image)
plt.show()
```

### Baboon

```python
baboon = images[0]
plt.imshow(baboon)
print_image_dimensions(baboon)
```

### Flowers

```python
flowers = images[1]
plt.imshow(flowers)
print_image_dimensions(flowers)
```

### Graffiti

```python
graffiti = images[2]
plt.imshow(graffiti)
print_image_dimensions(graffiti)
```

### Lena

```python
lena = images[3]
plt.imshow(lena)
print_image_dimensions(lena)
```

### Umbrella

```python
umbrella = images[4]
plt.imshow(umbrella)
print_image_dimensions(umbrella)
```

## Preprocessing the Images

```python
def describe_image_channels(image):
    for dimension in np.rollaxis(image, 2):
        print(pd.DataFrame(dimension).describe().round(2))
```

### Baboon

```python
baboon.shape
```

```python
describe_image_channels(baboon)
```

```python
# Baboon image is already normalized
baboon_reshaped = baboon.reshape(512*512, 3)
baboon_reshaped.shape
```

### Flowers

```python
flowers.shape
```

```python
describe_image_channels(flowers)
```

```python
# Flowers image is already normalized
flowers_reshaped = flowers.reshape(512*512, 4)
flowers_reshaped.shape
```

### Graffiti

```python
graffiti.shape
```

```python
pd.options.display.max_columns=7
describe_image_channels(graffiti)
```

```python
graffiti_normalized = graffiti / 255.0
```

```python
describe_image_channels(graffiti_normalized)
```

```python
graffiti_reshaped = graffiti_normalized.reshape(2048*2048, 3)
graffiti_reshaped.shape
```

### Lena

```python
lena.shape
```

```python
pd.options.display.max_columns=8
describe_image_channels(lena)
```

```python
# Lena image is already normalized
lena_reshaped = lena.reshape(512*512, 3)
lena_reshaped.shape
```

### Umbrella

```python
umbrella.shape
```

```python
pd.options.display.max_columns=7
describe_image_channels(umbrella)
```
